# import the dataset

```python
df = pd.read_csv("./data/train.csv", index_col = "observation_id")
df
```

```python
df.columns
```

```python
df.isna().sum()
```

Several columns with NaNs, most seems in unimportant columns which might not be used or can be replaced by False in case of binary

```python
# how many stations are we working with?
df.station.nunique(), df.station.unique()
```

```python
df["station"].value_counts().plot(kind="bar")
```

```python
# how many different outcomes we have?
df.Outcome.nunique(), df.Outcome.unique()
```

```python
df.Outcome.value_counts().plot(kind='bar')
plt.xlabel("Outcome")
plt.ylabel("Count")
```

```python
df.Outcome.value_counts(normalize = True)
```

There are many different Outcomes - ask which should be considered a success search? Looking to turn this into a binary problem, search or no search

```python
# how many age ranges we have?
df["Age range"].nunique(), df["Age range"].unique()
```

```python
df["Age range"].value_counts().plot(kind="bar")
plt.xlabel("Age");
plt.ylabel("Count");
```

```python
df["Age range"].value_counts(normalize=True)
```

There seems to be a weird age which is under 10, most likely to be dropped

```python
# Let's check on gender distribution
df["Gender"].value_counts().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Count")
```

```python
df["Gender"].value_counts(normalize=True)
```

Ooof the class looks very unbalanced when comparing Male to Female - Other might be dropped

```python
# Let's check on clothing removal info
df["Removal of more than just outer clothing"].value_counts().plot(kind="bar")
```

Again, a very high unbalance in both classes - even more if we convert the NaNs to False

There are a lot of NaN values around this class...hmm what to do? (426549/660611) -> 64,6%

```python
# and what gender does the NaNs most belong to?
df[df["Removal of more than just outer clothing"].isna()].Gender.value_counts().plot(kind="bar")
```

```python
df[df["Removal of more than just outer clothing"].isna()].Gender.value_counts(normalize=True)
```
